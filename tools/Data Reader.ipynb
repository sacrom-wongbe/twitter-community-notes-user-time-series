```python
import duckdb
import pandas as pd

path = r"C:\Users\wongb\twitter-community-notes-time-series\twitter-community-notes-user-time-series\aggregator\data\user_period_master_complete.parquet"

con = duckdb.connect()

# Read the parquet into DuckDB - get schema
df = con.execute(f"SELECT * FROM read_parquet('{path}') LIMIT 0").fetchdf()

# Show columns
print("Columns:")
print(df.columns)

# Get total row count
total_rows = con.execute(f"SELECT COUNT(*) FROM read_parquet('{path}')").fetchone()[0]
print(f"\nTotal Rows: {total_rows:,}")

# Get null counts for each column
null_counts = []
for col in df.columns:
    null_count = con.execute(f"""
        SELECT COUNT(*) - COUNT({col}) as null_count 
        FROM read_parquet('{path}')
    """).fetchone()[0]
    null_pct = (null_count / total_rows * 100) if total_rows > 0 else 0
    null_counts.append({
        'column': col,
        'null_count': null_count,
        'null_percentage': round(null_pct, 2)
    })

# Create null counts DataFrame
null_df = pd.DataFrame(null_counts)

# Display null counts
print("\nNull/NaN Counts by Column:")
pd.set_option('display.max_rows', None)
display(null_df)

# Get 10 random rows
print("\n10 Random Rows:")
random_rows = con.execute(f"SELECT * FROM read_parquet('{path}') USING SAMPLE 10").fetchdf()

# Set display options for better viewing
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)

# Display the DataFrame
display(random_rows)

con.close()
```
